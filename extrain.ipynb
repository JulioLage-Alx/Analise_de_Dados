```python
acimaMedia = Big[Big['Performance Index'] >= 55]['Sleep Hours'].mean()
abaixoMedia = Big[Big['Performance Index'] < 55]['Sleep Hours'].mean()
newdata = pd.DataFrame({'Alunos': ["Acima da mediana (x>=55)", "Abaixo da mediana (x<55)"], 'Tempo medio de sono': [acimaMedia, abaixoMedia]})
```

```python
ax = sns.barplot(data=newdata, x='Alunos', y='Tempo medio de sono')
ax.set_title("Análise da mediana")
```

```python
ax = sns.relplot(data=Big, x='Hours Studied', y='Performance Index', kind="line", errorbar=None,hue='Sleep Hours')
```
